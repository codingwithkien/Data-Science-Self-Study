```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv("california_cities.csv")
```

```python
cities.head(5)
```

```python
cities.info();
```

```python
latd, longd = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']
```

```python
plt.figure(figsize=(8,6))
plt.scatter(longd, latd, c=np.log10(population), s=area, cmap='viridis', linewidths=0, alpha=0.7);
plt.axis('equal')
plt.xlabel("Longlatitude")
plt.ylabel("Latdtitude")
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3,7);

# create a legend cities's size

area_range = [50, 100, 300, 500]
for area in area_range:
    plt.scatter([], [], s=area, c='k', label = str(area) + " km$^2$", alpha=0.4)
    
plt.legend(labelspacing=1, title="City Area")

plt.title("California Cities: Population and Area Distribution");
```
