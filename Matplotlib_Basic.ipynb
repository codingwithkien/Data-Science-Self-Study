## PyplotAPI vs Object Oriented (OO) API

### PyplotAPI

```python
x = np.linspace(0, 10, 10000)
plt.plot(x, np.sin(x), color = "blue", linestyle = "dashed", label = "Sin x");
plt.plot(x, np.cos(x), color = "red", label = "Cos x");

plt.title("Sin X and Cos X");
plt.xlabel("X");
plt.ylabel("Sin X");
# plt.xlim([0,8]); # x limit from 0 to 8

plt.legend(); # graph annotation
```

```python
x = np.linspace(0, 10, 10000)
plt.plot(x, np.sin(x), color = "blue", linestyle = "dashed", label = "Sin x");
plt.xlabel("X");
plt.ylabel("Sin X");
plt.xlim([0, 4]);
plt.ylim([0, 0.75]);
```

## Object Oriented API

```python
# prepare data
x = [1, 2, 3, 4, 5]
y = [11, 22, 33, 44, 55]

# set up plot
fig, ax = plt.subplots(figsize = (10, 5)) #

# plot the data
ax.plot(x, y);
ax.set(title = "A simple plot", xlabel = "x", ylabel = 'y');
```

#### scatter

```python
rng = np.random.RandomState(0) # unchange 

x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = 1000*rng.rand(100)

fig, ax = plt.subplots()
img = ax.scatter(x, y, s=sizes, c=colors, alpha=0.5);

fig.colorbar(img) # import color bar for graph
```

### Bar 

### Bar Vertical

```python
soft_dink_prices = {"Coca": 12, "Beer": 20, "Pessi": 10}
fig, ax = plt.subplots()
ax.bar(soft_dink_prices.keys(), soft_dink_prices.values(), color="green", alpha = 0.5);
ax.set(title = "Kien's Store", ylabel="Price ($)", xlabel = "Soft Drink");
```

```python
fig, ax = plt.subplots()
ax.barh(list(soft_dink_prices.keys()), list(soft_dink_prices.values()), color="green", alpha = 0.5);
```

```python
fix, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

ax1.plot(x, x * 2, color="red");
ax1.plot(x, x + 2, color="blue");

ax2.bar(soft_dink_prices.keys(), soft_dink_prices.values(), color="green", alpha = 0.5);
ax3.scatter(x, y, s=sizes, c=colors, alpha=0.5);
ax4.hist(student_height, alpha=0.5);
```
