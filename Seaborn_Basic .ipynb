# Introduction Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
print(sns.get_dataset_names())
```

## Numerical Plots

### Distribution Plot

```python
tips_df = sns.load_dataset('tips')
```

```python
tips_df.head()
```

```python
sns.histplot(data=tips_df['total_bill']);
```

 ### KDE Plot

```python
sns.kdeplot(data=tips_df['total_bill']);
```

### Displot

```python
sns.displot(data=tips_df, x='total_bill', col='time', kde=True);
```

### Bar Plot

```python
sns.barplot(data=tips_df, x='sex', y='tip', estimator=np.mean);
```

### Count Plot

```python
tips_df['sex'].value_counts()
```

```python
sns.countplot(data=tips_df, x='sex');
```

### Box plot

```python
sns.boxplot(data=tips_df, x='day', y='total_bill', hue='sex', palette="Blues");
plt.legend(loc=0);
```

### Facet Grid

```python
tips_fg = sns.FacetGrid(data=tips_df, row='smoker', col='time')
tips_fg.map(sns.scatterplot, 'total_bill', 'tip');
```

```python
tips_fg = sns.FacetGrid(data=tips_df, col='sex', hue='smoker', col_order=["Female", "Male"])
tips_fg.map(sns.scatterplot, 'total_bill', 'tip');
```

### Join Plot

```python
penguins_df = sns.load_dataset('penguins')
```

```python
penguins_df.head()
```

```python
sns.jointplot(data=penguins_df, x='flipper_length_mm', y='bill_depth_mm', hue='species');
```

### Pair Plot

```python
sns.pairplot(data=penguins_df, hue='species');
```

### HeatMaps

```python
flight_df = sns.load_dataset('flights')
```

```python
flight_df.head()
```

```python
flights = pd.pivot_table(flight_df, index='month', columns='year', values='passengers')
```

```python
flights
```

```python
sns.heatmap(flights, cmap="Blues", linecolor='white', linewidths=1);
```
